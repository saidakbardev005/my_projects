# Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/reklama_sales.csv')
df.head()
```

#TV & Sales

```python
TV=np.asanyarray(df["TV"]).reshape(-1,1)
Radio=df["Radio"]
Newspaper=df["Newspaper"]
Sales=np.asanyarray(df["Sales"]).reshape(-1,1)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Sales', y='TV')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='Sales', y='TV', line_kws={"color": "red"})
plt.show()
```

```python
LR_model = linear_model.LinearRegression()
LR_model.fit(TV,Sales)
```

```python
theta1 = LR_model.coef_[0][0]
theta0 = LR_model.intercept_[0]
print (f'{theta1=}\n'f'{theta0=}')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='TV', y='Sales')
x = np.linspace(df['TV'].min(), df['TV'].max(), 100)
plt.plot(x, theta1*x + theta0, '-r')
plt.xlabel("TV")
plt.ylabel("Sales")
```

```python
TV=np.asanyarray(df["TV"]).reshape(-1,1)
Sales=np.asanyarray(df["Sales"]).reshape(-1,1)
```

```python
y_predict = LR_model.predict(TV)
```

```python
MAE = mean_absolute_error(Sales, y_predict)
RMSE = np.sqrt(mean_squared_error(Sales, y_predict))
print(f"{MAE=}")
print(f"{RMSE=}")
```
